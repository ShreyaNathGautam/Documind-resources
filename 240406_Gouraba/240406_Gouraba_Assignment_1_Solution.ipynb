```python

daily_temperatures = [22, 25, 19, 28, 23, 30, 26]
temperature_threshold = 24

def analyze_temperatures(temperatures, threshold):
    filtered = [t for t in temperatures if t > threshold]
    mean = sum(filtered) / len(filtered) if filtered else 0
    return mean, filtered

mean_temp, filtered_temps = analyze_temperatures(
    daily_temperatures, temperature_threshold
)

print("Temperatures above threshold:", filtered_temps)
print("Average of temperatures above threshold:", mean_temp)
```

```python

import random

lists = [[random.randint(1, 100) for _ in range(5)] for _ in range(10)]

highest_sum = -1
highest_list = None

for i, lst in enumerate(lists, start=1):
    s = sum(lst)
    avg = s / len(lst)
    print(f"List {i}: {lst}")
    print("  Sum =", s)
    print("  Average =", avg)

    if s > highest_sum:
        highest_sum = s
        highest_list = lst

print("\nList with highest sum:", highest_list)
print("Highest sum:", highest_sum)
```

```python

import numpy as np
temps = np.array(daily_temperatures)
print("NumPy Array:", temps)
print("Mean:", np.mean(temps))
print("Maximum:", np.max(temps))
print("Minimum:", np.min(temps))
print("Standard Deviation:", np.std(temps))
```

```python

book_data = [
    {'title': '1984', 'author': 'George Orwell'},
    {'title': 'Brave New World', 'author': 'Aldous Huxley'},
    {'title': 'Animal Farm', 'author': 'George Orwell'},
    {'title': 'The Great Gatsby', 'author': 'F. Scott Fitzgerald'},
    {'title': 'To Kill a Mockingbird', 'author': 'Harper Lee'}
]

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

books = [Book(item['title'], item['author']) for item in book_data]

print("All Books:")
for book in books:
    print(f"{book.title} - {book.author}")

orwell_books = [book.title for book in books if book.author == "George Orwell"]

print("\nBooks by George Orwell:")
for title in orwell_books:
    print(title)

print("\nTotal number of books:", len(books))
```
